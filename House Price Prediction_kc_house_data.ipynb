# Training a Linear Regression Model

Split up the data into an X array that contains the features to train on, and a y array with the target variable whih is the Price to be predicted in this case. Skipping out the Date column because it has String info that the linear regression model can't use.

### X and y arrays

```python
#X = KC_Housing[['bedrooms', 'bathrooms', 'sqft_living',
#       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
#       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
#       'lat', 'long', 'sqft_living15', 'sqft_lot15']]
X = KC_Housing.drop(['id', 'date', 'price'], axis=1)
y = KC_Housing['price']
```

### Train Test Split 

Splitting the data into a training set and a testing set. Would be training out model on the training set and then use the test set to evaluate the model.
Taking 60% of Data for Training.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

### Creating and Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

### Model Evaluation

```python
# print the intercept
print(lm.intercept_)
```

```python
print(len(lm.coef_))
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

### Predictions from the Model

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

**Residual Histogram**

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('Variance score: %.3f' % r2_score(y_test, predictions))
```

# Inference

Variance score of approx 0.69 shows that 69% of total variation is described by our Linear model.
Which in turn means that 31% of the variation is not predicted by our model.
